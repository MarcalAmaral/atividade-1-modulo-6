####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')
plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2,2)
```

```python
sexo = sns.countplot(ax = axes[0,0], x = 'SEXO', data = df)
gestacao = sns.countplot(ax  = axes[0,1], x = 'GESTACAO', data = df)
estcivmae = sns.countplot(ax = axes[1,0], x = 'ESTCIVMAE', data = df)
consultas = sns.countplot(ax = axes[1,1], x = 'CONSULTAS', data = df)

plt.subplots_adjust(wspace=0.5, hspace=1)
```

```python
labels = gestacao.get_xticklabels()
gestacao.set_xticklabels(labels, rotation = 60)
```
